## Preparing example data

```python
features = [
    'CD45RA', 'CD133', 'CD19', 'CD22', 'CD11b', 'CD4', 'CD8', 'CD34',
    'Flt3', 'CD20', 'CXCR4', 'CD235ab', 'CD45', 'CD123', 'CD321',
    'CD14', 'CD33', 'CD47', 'CD11c', 'CD7', 'CD15', 'CD16', 'CD44',
    'CD38', 'CD13', 'CD3', 'CD61', 'CD117', 'CD49d', 'HLA-DR', 'CD64', 'CD41'
]
```

```python
# Read the Levine32 FCS file from disk
# By default, CytoTools reads data in as Polars dataframe, but we can easily convert to a Pandas DataFrame
data = polars_to_pandas(read_from_disk("data/Levine_32dim_notransform.fcs"))

# Replace the labels with more intuitive population names
populations = pd.read_csv("data/levine32_populations.csv")
mappings = {x["label"]: x["population"] for x in populations.to_dict("records")}
data["label"] = data["label"].replace(mappings)
data.rename(columns={"label": "Ground Truth", "individual": "sample_id"}, inplace=True)

# Remove events that do not have a ground truth label.
data = data[~data["Ground Truth"].isnull()].copy()

reducer = DimensionReduction(method="UMAP", n_components=2, n_jobs=-1, n_neighbors=30)
data_with_embeddings = apply_transform(data=data.copy(), features=features, method="asinh", cofactor=100)
data_with_embeddings = reducer.fit_transform(data=data_with_embeddings, features=features)
```

```python
data_with_embeddings.to_csv("data/cache.csv", index=False)
```

```python
data_with_embeddings = pd.read_csv("data/cache.csv")
```

## Base clustering

```python
clustering = SingleClustering.from_dataframe(
    data=data_with_embeddings,
    features=features,
    verbose=True,
    random_state=42,
    pre_embedded=True
)
```

```python
clustering.global_clustering(
    method="flowsom",
    clustering_params=dict(
        som_dim=(10, 10), learning_rate=0.2, batch_size=100
    )
)
```

```python
clustering.plot_cluster_membership(
    include_legend=True, 
    legend_kwargs=dict(bbox_to_anchor=(1.0, 1.0))
)
```

```python
clustering.data.cluster_label.to_csv("data/labels/flowsom.csv", index=False)
```

```python
clustering.plot_cluster_membership(
    include_legend=True, 
    legend_kwargs=dict(bbox_to_anchor=(1.0, 1.0))
)
```

```python
clustering.data.cluster_label.to_csv("data/labels/phate_kmeans.csv", index=False)
```

## Ensemble clustering with graph based methods

```python
labels = {
    x.replace(".csv", ""): pd.read_csv(f"data/labels/{x}").cluster_label.values
    for x in os.listdir("data/labels") if x != ".ipynb_checkpoints"
}
```

```python
clustering = EnsembleClustering.from_dataframe(
    data=data_with_embeddings,
    labels=labels,
    features=features,
    verbose=True,
    random_state=42,
    pre_embedded=True
)
```

```python
results = clustering.k_performance(k_range=(5, 15), consensus_method="mcla", sample_n=1000, resamples=100)
```

```python
from cytoplots.general import box_swarm_plot, ColumnWrapFigure

fig = ColumnWrapFigure(n=4, col_wrap=2, figsize=(8, 8))
for metric, df in results.groupby("Metric"):
    ax = fig.add_wrapped_subplot()
    box_swarm_plot(
        data=df,
        x="K",
        y="Value",
        ax=ax
    )
    ax.set_title(metric)
fig.tight_layout()
fig
```

```python
clustering.graph_consensus_clustering(consensus_method="mcla", k=6)
```

```python
clustering.plot_cluster_membership(
    include_legend=True, 
    legend_kwargs=dict(bbox_to_anchor=(1.0, 1.0))
)
```

## Ensemble clustering with GeoWave

```python
clustering.geometric_median_heatmap(
    method="average", 
    metric="cityblock", 
    cmap="coolwarm", 
    xticklabels=True,
    yticklabels=True,
    plot_orientation="horizontal",
)
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering.geowave(
    method="k_consensus", 
    method_kwargs=dict(
        smallest_cluster_n=10, 
        largest_cluster_n=15,
        n_resamples=100,
        clustering_klass=AgglomerativeClustering(affinity="manhattan", linkage="average"),
    )
)
```

```python
clustering.plot_cluster_membership(
    include_legend=True, 
    legend_kwargs=dict(bbox_to_anchor=(1.0, 1.0))
)
```
